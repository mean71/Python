# 1.시리즈 (Series)

1) Pandas Data
- 시리즈(Series); 데이터가 순차적으로 나열된 1차원 배열의 형태
- 딕셔너리(dictionary)와 유사한 구조: 키(key)와 값(value)
- pd.Series (dictionary_type_data)
- 정수형 위치 인덱스/인덱스 이름(인덱스 라벨)
- Series.index
- Series.values  

1차원 배열과 유사한 자료 구조  
데이터와 인덱스(index)가 함께저장.  
각 데이터는 인덱스를 통해 접근  
인덱스는 기본적으로 0부터 시작, 사용자가 원하는 값으로 설정할 수도  
다양한 데이터 타입을 저장, 데이터분석에 많이 사용  
1차원 데이터 구조이기 때문에  
데이터프레임의 각 행/열은 본질적으로 인덱스와 값의 쌍으로 시리즈가 될 수 있다

s = pd.Series(data, index=index, dtype=dtype, name=name, copy=none)  
 주요 파라미터:  
- data: 리스트,튜플,딕셔너리,스칼라값 등  
예:pd.Series([1,2,3,4]) 또는 pd.Series({'a':1,'b':2})  
- index:Series의 라벨 지정  
예:pd.Series([1,2,3]), index=['a','b','c']  
- dtype:데이터 타입 지정(기본값은 자동 추론)  
예: pd.Series([1,2,3]), dtype='float64'  
- name:pd.Series의 이름 지정  
예: pd.Series([1,2,3])name='numbers'  
- copy: 입력 데이터의 복사본 생성 여부(기본값 False)

```python
# 데이터프레임을 병합할 때는 pd.concat() 함수를 사용(기본이 세로(위+아래)로 연결)
import pandas as pd
# 두 개의 데이터프레임 생성
df1 = pd.DataFrame({'A': [1,2,3], 'B':[4,5,6]})
df2 = pd.DataFrame({'A': [7,8,9], 'B':[10,11,12]})

# 데이터프레임 병합(세로로 결합)
combined_df = pd.concat([df1, df2], ignore_index=True)

#df1
#df2
combined_df
```

```python
# 데이터프레임을 병합할 때는 pd.concat()함수를 사용(가로연결: axis = 1)
import pandas as pd
# 두 개의 데이터프레임 생성
df1 = pd.DataFrame({'A': [1,2,3], 'B':[4,5,6]})
df2 = pd.DataFrame({'A': [7,8,9], 'B':[10,11,12]})
#데이터프레임 병합(가로로 결합)
merged_df =  pd.concat([df1, df2], axis=1)
print(merged_df)
```

- 데이터프레임을 병합 : join()  

    두개의 데이터프레임을 병합할 때 인덱스를 기준으로 데이터 프레임을 결합.  

DataFrame.join(other, on=None, how='left', lsuffix='', rsuffix='', sort=False)  
  - other : 병합할 다른 데이터프레임  
  - on : 인덱스가 아닌 컬럼을 기준으로 병합할 경우 사용  
  - how : 병합 방식 ('left','right','outer',- 'inner')  
  - lsuffix, rsuffix: 병합할 때 컬럼 이름이 중복될 경우 접두사  
  - sort : 병합 후 정렬 여부  

1. 'left'  
왼쪽 데이터프레임(df1)의 모든 행을 유지하며 우측 데이터프레임(df2)의 일치하는 데이터를 병합.  
만약 우측에 일치하는 데이터가 없다면 NaN으로  
2. 'right'  
우측 데이터프레임(df2)의 모든 행을 유지하며 왼쪽 데이터프레임(df1)의 일치하는 데이터를 병합.  
만약 좌측에 일치하는 데이터가 없다면 NaN으로  
3. 'outer'  
두 데이터프레임(df1과 df2)의 모든 행을 유지하며 병합  
일치하지 않는 데이터는 NaN으로  
4. 'inner'  
두 데이터프레임(df1과 df2)에서 공통으로 존재하는 행만 유지하며 병합  
일치하지 않는 데이터는 제외

```python
# 데이터프레임을 병합: join(),outer(모든 행을 유지)
import pandas as pd
#예제 데이터프레임 생성
df1 = pd.DataFrame({'A': ['A0','A1','A2'],
                    'B': ['B0','B1','B2']},
                   index=['K0','K1','K2'])
df2 = pd.DataFrame({'C': ['C0','C1','C2'],
                    'D': ['D0','D1','D2']},
                   index=['K0','K2','K3'])
#join() 사용
result =  df1.join(df2, how='outer')
print(result)
```

### merge()

컬럼을 기준으로 병합하는데 유용  

pandas.merge(left,right,how='inner',on=None,left_on=None,right_on=None,left_index=False,right_index=False,sort=False)  

  left:왼쪽 데이터프레임  
  right:오른쪽 데이터프레임  
  how:병합 방식('left','right','outer','inner')  
  on:병합 기준이 되는 컬럼 이름  
  left_on,right_on:병합 기준이 되는 왼쪽 및 오른쪽 데이터프레임의 컬럼 이름  
  left_index,right_index:인덱스를 기준으로 병합할지 여부  
  sort:병합 후 정렬 여부  

```python
# 데이터프레임을 병합:merge(),inner&key(키값이 일치하는 행만!)
import pandas as pd
# 예제 데이터프레임 생성
df1 = pd.DataFrame({
   'key':['k0','k1','k2','k3'],
   'A': ['A0','A1','A2','3'],
   'B': ['B0','B1','B2','B3']})
df2 = pd.DataFrame({
   'key':['k0','k1','k2','k4'],
   'C': ['C0','C1','C2','C3'],
   'D': ['D0','D1','D2','D3']})
# merge() 사용
result = pd.merge(df1,df2, on='key',how='inner')
print(result)
```

## 2.3.DataFrame 행 데이터 선택

- iloc(integer position based):  
정수 위치 인덱스를 사용하여 행 데이터를 지정  
0부터 시작하는 정수 인덱스로 마지막을 불포함  
-loc(label based location):  
인덱스 이름(라벨)을 기준으로 행 데이터를 지정  
라벨을 기반으로 시작하여 끝부분을 포함  

```python
import pandas as pd

df = pd.DataFrame({
    '이름': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    '나이': [25,30,35,28,22],
    '도시': ['서울','부산','인천','대구','광주']
}, index=['a','b','c','d','e'])
print(df)
```

```python
#loc(label based location):인덱스 이름(라벨:abcde)을 기준으로 행 데이터를 지정
#인덱스 'b'인 행 선택
print(df.loc['b'])
#인덱스 'b'부터 'd'까지의 행과 '이름','나이' 열 선택
print(df.loc['b':'d',['이름','나이']])
#조건을 사용한 선택
print(df.loc[df['나이'] > 30])
```

```python
#iloc(integer position based):정수 위치 인덱스(0~4)를 사용하여 행 데이터를 지정
# 두 번째 행 선택 (인덱스 1)
print(df.iloc[1])
# 1부터 3까지의 행(1,2번째 행)과 0, 2번째 열 선택
print(df.iloc[1:3, [0,2]])
# 처음 3개 행과 모든 열 선택
```

1. 단일 열 선택 :

  점(.) 표기법(공백이 없는 열 이름에만 사용 가능):df.column_name  
2. 여러 열 선택 :  
  리스트를 사용한 대괄호 표기법 : df[['column1','column2','column3']]  
  조건을 이용한 열 선택 : df.loc[:,df.columns.str.contains('pattern')]  
3. 인덱스 기반 선택(iloc) :  
  df.iloc[:,0] # 첫 번째 열  
  df.iloc[:,[0,2,4]] # 첫 번째, 세 번째, 다섯 번째 열  
  df.iloc[:,0:3] # 처음 세 개의 열  
- 열데이터 선택
1. 라벨 기반 선택(loc) :  
  df.loc[:,'column_name']  
  df.loc[:,['column1','column2']]  
2. 이름함수를 이용한 선택 :  
  df.filter(items=['column1','column2'])  
3. 불리언 인덱싱:  
  mask=[True,False,True,False,...]  
  --DataFrame의 열 수 만큼 df.loc[:,mask]

```python
import pandas as pd
#샘플 DataFrame 생성
df = pd.DataFrame({
    'A':[1,2,3],
    'B':[4,5,6],
    'C':[7,8,9],
    'D':[10,11,12]
})

#단일 열 선택
print(df['A'])
print(df.B)
#여러 열 선택
print(df[['A','C']])
#조건을 이용한 열 선택
print(df.loc[:,df.columns.str.contains('A|B')])
#iloc를 이용한 선택
print(df.iloc[:,[0,2]])
#loc를 이용한 선택
print(df.loc[:,['A','D']])
#데이터 타입 기반 선택
print(df.select_dtypes(include=['int64']))
```

## --Groupby()

데이터를 특정 기준으로 그룹화  
각 그룹에 대해 다양한 연산수행  

그룹화 기준이 되는 열의 이름.  
그룹화된 객체에 대해 집계 함수(sum(),mean(),count()등)를 호출

```python
import pandas as pd
#예제 데이터프레임 생성
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Value': [10, 20, 30, 40, 50, 60]
}
df = pd.DataFrame(data)
# 'Category'열을 기준으로 그룹화하고, 그룹별 합계 계산
grouped = df.groupby('Category').sum()
print(grouped)
```

```python
import pandas as pd
# 예제 데이터프레임 생성
data = {
    'Category1': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Category2': ['X','Y','X','Y','X','Y'],
    'Value': [10, 20, 30, 40, 50, 60]
}
df = pd.DataFrame(data)
# 'Category1'과 'Category2' 열을 기준으로 그룹화하고, 그룹별 합계 계산
grouped = df.groupby(['Category1', 'Category2']).sum()
print(grouped)
```

```python
import pandas as pd
#예제 데이터프레임 생성
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Value': [10, 20, 30, 40, 50, 60]
}
df = pd.DataFrame(data)
# 'Category'열을 기준으로 그룹화하고, 그룹별 평균 계산
grouped = df.groupby('Category').agg(['sum','mean'])
print(grouped)
```

```python
import pandas as pd
# 예제 데이터 프레임 생성
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Values': [10, 20, 30, 40, 50, 60]
}
df = pd.DataFrame(data)
# 사용자 정의 함수: 그룹의 최대값과 최소값의 차이 계산
def range_func(x):
  return x.max() - x.min()
# 'Category' 열을 기준으로 그룹화하고, 사용자 정의 함수 적용
grouped = df.groupby('Category').apply(lambda x: range_func(x['Values']))
print(grouped)
```

```python
import pandas as pd
# 예제 데이터프레임 생성
data = {
    'Category': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Values': [10, 20, 30, 40, 50, 60]
}
df = pd.DataFrame(data)
# 'Category'열을 기준으로 그룹화하고, 그룹의 한계가 50 이상인 그룹만 선택
filteed = df.groupby('Category').filter(lambda x: x['Values'].sum() > 50)
print(filteed)
```

### ___Agg()

데이터프레임/시리즈에 대해 여러 집계연산 동시수행  
groupby()함수와 함께 사용하면 그룹별 집계연산 쉽게수행  
sum:합계, mean:평균, median:중앙값, min:최솟값, max:최대값,count:값의 개수, std:표준편차, var:분산, prod:곱

```python
df = pd.DataFrame({
    'Category': ['A', 'A', 'B', 'B'],
    'Values1': [1, 2, 3, 4],
    'Values2': [10, 20, 30, 4]
    })
grouped = df.groupby('Category')
# 그룹별로 여러 집계 함수 적용
result = grouped.agg({'Values1': ['sum', 'mean'], 'Values2': 'max'})
print(result)
```

- dataFrame.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)  
1. axis: 결측값을 기준으로 제거할 축을 지정. 0은 행을, 1은 열을 의미. 기본값은 0입  
2. how:'any'또는 'all. 'any'는 하나라도 NaN이 있으면 해당 행 또는 열을 제거하고, 'all'은 모든 값이 NaN일 때만 제거. 기본값은 'any'  
3.thresh:이 값보다 많은 비결측값이 있는 행 또는 열만 남긴다.  
4. subset:특정 행이나 열에서만 결측값을 찾아 제거  
5. inplace: True로 설정하면 원본 데이터프레임을 수정

```python
# 데이터 프레임의 결측치 삭제: dropna()
import pandas as pd
import numpy as np
# 예제 데이터프레임 생성
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [np.nan, 2, 3, 4],
    'C': [1, 2, 3, np.nan,]
})
print('원본 데이터프레임:\n', df)
#결측값이 있는 행 제거
df_dropped = df.dropna()
print('결측값이 있는 행 제거:\n', df_dropped)
# 모든 값이 NaN인 행 제거
df_dropped_all = df.dropna(how='all')
print('\n모든 값이 NaN인 행 제거:\n', df_dropped_all)
# 특정 열에서만 결측값이 있는 행 제거
df_dropped_subset = df.dropna(subset=['A','B'])
print('특정 열에서만 결측값이 있는 행 제거:\n', df_dropped_subset)
```

## --데이터프레임 결측값 채우기 : fillna()

데이터프레임에서 결측값(NaN)을 지정값으로 대체  

DataFrame.fillna(value=None, method=None, exis=None, inplace=False, limit=None)  
- value: 결측값을 대체할 값. 딕셔너리,시리즈,스칼라 값 등을 사용가능  
- method: 'ffill' (앞의 값으로 채우기) 또는 'bfill'(뒤의 값으로 채우기)  
- axis: 결측값 채울 축. 0 또는 'index'는 행따라, 1 또는 'columns'는 열따라.  
- inplace: 원본데이터프레임 변경여부
- limit: 결측값을 채울 최대 개수

```python
import pandas as pd
import numpy as np
# 예제 데이터 프레임 생성
df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [np.nan, 2, 3, np.nan, 5],
    'C': [1, np.nan, np.nan, 4, 5],
    'D': [np.nan, np.nan, np.nan, np.nan, np.nan]
})
print('원본 데이터 프레임:\n',df)
```

## --bining(구간화)

- 연속형데이터를 여러구간(bin)으로 나누는 과정  
- 데이터분포를 분석이나 시각화할 때 유용  
- pandas에선 주로 cut()과 qcut()함수를 사용하여 binning수행  

"Bin" 통,상자,저장소 의미  
binning은 연속적인 데이터를 이산적인 구간으로 나누는 과정  
연속형 변수를 범주형 변수로 변환

### ___cut()함수

데이터를 지정구간(bin)으로 나눕니다.  
비율이 아닌 연속적인 값의 범위에 따라 구간을 나누는 데 사용

```python
import pandas as pd
# 예제 데이터
data = [1,7,5,4,6,3,8,9,2,10]
# 데이터프레임 생성
df = pd.DataFrame(data, columns=['Value'])
# 구간 정의
bins = [0,3,6,9,12]
# cut() 함수를 사용하여 데이터 binning
df['Binned']= pd.cut(df['Value'], bins)
print(df)
```

### ___qcut()

데이터를 동일한 개수의 관측치가 들어가도록 구간(bin)으로 분할  
각 구간에 들어가는 데이터개수가 거의 동일하도록 분할

```python
import pandas as pd
# 예제 데이터
data = [1,7,5,4,6,3,8,9,2,10]
# 데이터프레임 생성
df = pd.DataFrame(data, columns=['Value'])
# qcut()함수를 사용하여 데이터 binning (4개의 구간으로 나눔)
df['Binned'] = pd.qcut(df['Value'], 4)
print(df)
```

```python
# 데이터프레임 생성
food_1 = pd.DataFrame({
    'name': ['햄버거', '똥', '라면', '엽떡', '탕후루', '마라탕', '피자'],
    'healthy-rate': [1, 2, 2, 3, 1, 2, 1],
}, index=[1,2,3,4,5,6,7])

food_2 = pd.DataFrame({
    'name': ['마늘', '양파', '도라지', '히비스커스', '볼펜', '샤프심', '지우개똥'],
    'healthy-rate': [10, 12, 13, 21, 15, 30, 23],
}, index=[1,2,3,4,5,10,7])

df = pd.concat([food_1, food_2])

grouped = df.groupby('name')['healthy-rate'].mean().sort_values(ascending=False).reset_index()
grouped

top = grouped.iloc[:5].sort_values(by='healthy-rate', ascending=False)
bottom = grouped.iloc[-5:].sort_values(by='healthy-rate')

print("건강한 음식 Top 5:")
print(top)

print("\n가장 건강하지 않은 음식 Bottom 5:")
print(bottom)
```
