````python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd

plt.rc("font", family = "gulim")
plt.rc("axes", unicode_minus = False) # 마이너스 출력할 때 에러나지 않도록
````

````python
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y = x*0.1
y2 = np.sin(x)
y3 = np.cos(x)
y4 = np.tan(x)

plt.figure(num=1, figsize=(10,5))
# , frameon=True, constrained_layout=True
# 그림 테두리 표시 여부 (True/False) # 더 세밀한 여백 레이아웃 조정 (True/False)


y_list = [y, y2, y3, y4]
y_label = ['y', 'y2', 'y3', 'y4']
color_list = ['b', 'g', 'r']
fig, axis = plt.subplots(3, 1, figsize = (12,10))

for i in range(3):
  axis[i].plot(x, y_list[i], label = y_label[i], c = color_list[i])
  axis[i].set_ylim(-1, 1)
  axis[i].grid(linestyle = ':')
  axis[i].set_xticks([-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], labels = ['-2pi', '-pi', '0', 'pi', '2pi'])


plt.plot( # plot() 함수의 위치/키워드 인수 옵션들
# 위치인수
    x,y, # x,y축 값 (데이터 리스트 또는 배열)
    '-.>r', # 형식 문자열 'fmt' (선 스타일, 마커, 색상) - 생략 가능.
    # 선스타일 : (-/--/-./:) (실선 / 점선 / 대시-점선 / 점선)
    # 마커스타일 : o/s/D/x/^/v/</>/p/h/+/1/2/3/4 (원,사각,다이아,엑스,상하좌우삼각,오각,육각,+,상하좌우)
    # 색상 : b/g/r/c/m/y/k/#RRGGBB (파b/초g/빨r/청c/자홍m/노y/검k/#16진수색상코드)
# 키워드 인수
    color = None, # 선의 색상. (예, # 'r', 'blue', '#FF5733')
    linestyle = None, # 선의 스타일. (예, # '-' 실선, '--' 점선, '-.' 대시-점, ', #' 점선)
    linewidth = 1, # 선의 두께. (기본값, # 1.5)

    marker = 'o', # 마커스타일. (예, # 'o', 's', 'D' '*')
    markersize = 4, # 마커크기 (기본값: 6)
    markeredgecolor = None, # 마커 테두리 색
    markeredgewidth = None, # 마커 테두리 두께
    markerfacecolor = None, # 마커 내부 색
    
    label = 'label =그래프 범례 텍스트',
    drawstyle = 'default', # 포인트 연결선 형식 dafault/steps/steps-pre/steps-post (연속적인 선/ 직각/ 이전포인트계단/ 이후포인트계단)
    
    aa = False, # antialiased = (True/False) : 라인 안티앨리어싱 처리 여부
    alpha = None, # 선 투명도 (투명 0 : 1 불투명(defalt))
    zorder = None, # 그래프 겹침 순서 (정수: 클수록 표시 우선)
)

plt.title('Polynomial curve') # 플롯 제목
plt.xlabel("X") # x축 라벨
plt.ylabel(
    "f(x)",
    rotation=0, # 라벨 축 각도
    fontsize=14, # 폰트크기
    labelpad=30 # 축-라벨사이 거리
) 
plt.legend( # 범례추가
    loc='upper right',     # 범례 위치 best / (upper/center/lower) (left/center/right)
    fontsize=12,           # 폰트 크기
    title='범례제목',
    ncol=2,                # 범례 N열로 정렬
    frameon=True          # 테두리
)

# 세로선 그리기
plt.vlines(x=0, ymin=-1, ymax=1, color='r')
# 가로선 그리기
plt.hlines(
    y=0,                # y축위치(y=0에 가로선)
    xmin=-2*np.pi,      # 선의 시작x축 값
    xmax=2*np.pi,       # 선의 끝x축 값
    color='k'           # 선 색상
)
plt.xlim(-2*np.pi, 2*np.pi) # x축 범위
plt.ylim(-1, 1) # y축 범위
# x축 눈금 설정
plt.xticks(
    ticks=[-2*np.pi, -np.pi, 0, np.pi, 2*np.pi], # x축 눈금 위치 (여기서는 -2π, -π, 0, π, 2π 위치에 눈금 설정)
    labels=['-2pi', '-pi', '0', 'pi', '2pi']    # 각 눈금의 레이블 (여기서는 -2pi, -pi, 0, pi, 2pi로 레이블 설정)
)

plt.grid(linestyle = ':') # 그리드 추가 (- / -- / :)
plt.show() #모든 플롯 출력
````

# Seaborn

matplotlib을 기반으로 구축된 고급 데이터 시각화 기능 라이브러리

### 주요 모듈 및 메서드

1. **`seaborn` 모듈**

   - **`seaborn`**: 기본 모듈로, 시각화에 사용되는 다양한 기능을 제공합니다.
   - **`seaborn.relplot()`**: 관계형 데이터 플롯 (scatter, line) 산점도 또는 선 그래프를 생성
   - **`seaborn.catplot()`**: 범주형 데이터 플롯 (bar, point, box) 범주형 데이터의 시각화를 위한 다양한 그래프를 생성합니다.

   - **`seaborn.histplot()`**: 히스토그램 생성
   - **`seaborn.kdeplot()`**: 밀도 추정 그래프를 생성합니다.
   - **`seaborn.boxplot()`**: 박스 플롯을 생성합니다.
   - **`seaborn.violinplot()`**: 바이올린 플롯을 생성합니다.
   - **`seaborn.pairplot()`**: 여러 변수 간의 관계를 시각화하는 플롯을 생성합니다. # 변수 간 관계 시각화
   - **`seaborn.heatmap()`**: 히트맵을 생성합니다.
   - **`seaborn.jointplot()`**: 두 변수 간의 관계를 시각화합니다. # 두 변수 간 관계와 분포를 동시에 표현

2. **주요 함수 및 메서드**

   - **`sns.scatterplot()`**: 산점도 그래프를 생성합니다.
   - **`sns.lineplot()`**: 선 그래프를 생성합니다.
   - **`sns.barplot()`**: 막대 그래프를 생성합니다.
   - **`sns.boxplot()`**: 박스 플롯을 생성합니다.
   - **`sns.histplot()`**: 히스토그램을 생성합니다.
   - **`sns.kdeplot()`**: 커널 밀도 추정 그래프를 생성합니다.
   - **`sns.violinplot()`**: 바이올린 플롯을 생성합니다.
   - **`sns.heatmap()`**: 히트맵을 생성합니다.
   - **`sns.pairplot()`**: 여러 변수 간의 관계를 시각화합니다.
   - **`sns.jointplot()`**: 두 변수 간의 관계를 시각화합니다.
   - **`sns.lmplot()`**: 회귀선이 포함된 산점도 그래프를 생성합니다. # 선형 회귀선과 함께 데이터의 관계 시각화

3. **`seaborn`의 주요 인수 및 옵션**

   - **`x`, `y`**: x축과 y축 데이터.
   - **`hue`**: 데이터의 색상 구분 변수.
   - **`size`**: 마커의 크기 구분 변수.
   - **`style`**: 마커의 스타일 구분 변수.
   - **`data`**: 데이터셋 (Pandas DataFrame 등).
   - **`palette`**: 색상 팔레트.
   - **`markers`**: 마커 스타일 사용 여부.
   - **`sizes`**: 마커 크기 설정.
   - **`style_order`**: 마커 스타일의 순서.
   - **`hue_order`**: 색상 순서.
   - **`x_jitter`, `y_jitter`**: 데이터에 임의의 노이즈 추가.
   - **`legend`**: 범례 표시 여부.
   - **`alpha`**: 투명도.
   - **`linewidth`**: 마커의 선 두께.
   - **`edgecolor`**: 마커의 테두리 색상.

4. **서브플롯 및 레이아웃**

   - **`sns.FacetGrid`**: 여러 플롯을 그리기 위한 그리드 객체.
   - **`sns.PairGrid`**: 변수 쌍 간의 관계를 시각화하는 그리드 객체.

5. **시각화 스타일**

   - **`sns.set()`**: 기본 시각화 스타일을 설정합니다.
   - **`sns.set_style()`**: 스타일을 설정합니다 (예: 'white', 'dark', 'whitegrid', 'darkgrid').
   - **`sns.set_palette()`**: 색상 팔레트를 설정합니다.

### 사용 예시

```python




# 산점도 생성
sns.scatterplot(x='total_bill', y='tip', data=df, hue='day', style='time')

# 히스토그램 생성
sns.histplot(df['total_bill'], kde=True)

# 박스 플롯 생성
sns.boxplot(x='day', y='total_bill', data=df)

# 히트맵 생성

````python
import seaborn as sns
import matplotlib.pyplot as plt
# 1. 주요 상수
# Seaborn의 스타일 옵션: 'darkgrid', 'whitegrid', 'dark', 'white', 'ticks'
sns.set_style("whitegrid")  # 배경 스타일 설정
# Seaborn의 색상 팔레트
# Seaborn은 다양한 색상 팔레트를 지원하며, 기본 팔레트와 커스터마이즈된 팔레트를 제공
palette = sns.color_palette("husl", 8)  # 'husl' 팔레트 사용, 8개의 색상
# 2. 주요 메서드

sns.get_dataset_names()  # 내장 데이터셋 이름을 리스트로 반환
data = sns.load_dataset("iris")  # iris 데이터셋 로드
# 2.3. scatterplot: 산점도 플롯
# 주요 인수: x, y, hue, style, size, palette, markers, data
sns.scatterplot(
    data=data, 
    x="sepal_length", 
    y="sepal_width", 
    hue="species",  # 색상별 그룹화
    palette=palette,  # 색상 팔레트
    style="species",  # 스타일별 그룹화
    size="species"  # 크기별 그룹화
)
plt.title("Scatterplot of Sepal Length vs Sepal Width")  # 플롯 제목 추가
plt.show()  # 플롯 출력

# 2.4. lineplot: 선 그래프
# 주요 인수: x, y, hue, style, size, palette, data
sns.lineplot(
    data=data, 
    x="sepal_length", 
    y="sepal_width", 
    hue="species",  # 색상별 그룹화
    palette=palette,  # 색상 팔레트
    style="species",  # 스타일별 그룹화
    size="species"  # 크기별 그룹화
)
plt.title("Lineplot of Sepal Length vs Sepal Width")  # 플롯 제목 추가
plt.show()  # 플롯 출력

# 2.5. barplot: 막대 그래프
# 주요 인수: x, y, hue, estimator, ci, palette, data
sns.barplot(
    data=data, 
    x="species", 
    y="sepal_length", 
    hue="species",  # 색상별 그룹화
    palette=palette,  # 색상 팔레트
    estimator=sum,  # 집계 함수 (기본값은 평균)
    ci=None  # 신뢰 구간 표시 여부
)
plt.title("Barplot of Sepal Length by Species")  # 플롯 제목 추가
plt.show()  # 플롯 출력

# 2.6. histplot: 히스토그램
# 주요 인수: x, y, hue, multiple, bins, kde, palette, data
sns.histplot(
    data=data, 
    x="sepal_length", 
    hue="species",  # 색상별 그룹화
    palette=palette,  # 색상 팔레트
    multiple="stack",  # 히스토그램을 스택 형식으로 표시
    bins=20,  # 히스토그램의 빈 개수
    kde=True  # 커널 밀도 추정 추가
)
plt.title("Histogram of Sepal Length")  # 플롯 제목 추가
plt.show()  # 플롯 출력

# 2.7. pairplot: 페어 플롯
# 주요 인수: hue, palette, markers, diag_kind, plot_kws
sns.pairplot(
    data, 
    hue="species",  # 색상별 그룹화
    palette=palette,  # 색상 팔레트
    markers=["o", "s", "D"],  # 마커 스타일
    diag_kind="kde"  # 대각선 그래프 유형 ('hist' 또는 'kde')
)
plt.title("Pairplot of Iris Dataset")  # 플롯 제목 추가
plt.show()  # 플롯 출력

# 2.8. heatmap: 히트맵
# 주요 인수: data, annot, cmap, center, linewidths, linecolor
sns.heatmap(
    data.corr(),  # 상관 행렬
    annot=True,  # 셀에 값 주석 추가
    cmap="coolwarm",  # 색상 맵
    center=0,  # 중심값 설정
    linewidths=0.5,  # 셀 간의 선 너비
    linecolor='gray'  # 셀 간의 선 색상
)
plt.title("Heatmap of Correlation Matrix")  # 플롯 제목 추가
plt.show()  # 플롯 출력

# 2.9. lmplot: 회귀 플롯
# 주요 인수: x, y, hue, palette, markers, scatter, line_kws
sns.lmplot(
    data=data, 
    x="sepal_length", 
    y="sepal_width", 
    hue="species",  # 색상별 그룹화
    palette=palette,  # 색상 팔레트
    markers=["o", "s", "D"],  # 마커 스타일
    scatter=True,  # 산점도 표시 여부
    line_kws={"color": "red", "linewidth": 2}  # 회귀선 색상 및 두께
)
plt.title("Regression Plot of Sepal Length vs Sepal Width")  # 플롯 제목 추가
plt.show()  # 플롯 출력

# 2.10. violinplot: 바이올린 플롯
# 주요 인수: x, y, hue, palette, scale, inner
sns.violinplot(
    data=data, 
    x="species", 
    y="sepal_length", 
    hue="species",  # 색상별 그룹화
    palette=palette,  # 색상 팔레트
    scale="area",  # 스케일 옵션 ('area', 'count', 'width')
    inner="quartile"  # 내부 표시 ('box', 'quartile', 'point', 'stick')
)
plt.title("Violin Plot of Sepal Length by Species")  # 플롯 제목 추가
plt.show()  # 플롯 출력

# 2.11. boxplot: 박스 플롯
# 주요 인수: x, y, hue, palette, showfliers, notch
sns.boxplot(
    data=data, 
    x="species", 
    y="sepal_length", 
    hue="species",  # 색상별 그룹화
    palette=palette,  # 색상 팔레트
    showfliers=True,  # 이상치 표시 여부
    notch=False  # 박스의 노치 여부
)
plt.title("Box Plot of Sepal Length by Species")  # 플롯 제목 추가
plt.show()  # 플롯 출력
````
